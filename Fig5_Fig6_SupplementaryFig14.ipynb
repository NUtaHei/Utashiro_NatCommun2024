```python
#import
import pickle
import glob
import seaborn as sns
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
from scipy import stats
from scipy.ndimage import gaussian_filter1d
```

```python
#load data
folder = '/Volumes/Elements/data_for_uploaded/'

all_Fres = np.load(folder + 'Fig.5/' + 'responsive_dFovF.npy', allow_pickle=True) 
df = pd.read_csv(folder + 'Fig.5/' + 'df_forFig5.csv') 
df2 = pd.read_csv(folder + 'Fig.5/' + 'df2_forFig5.csv') 
pos = np.load(folder + 'Fig.5/pos_labels.npy', allow_pickle=True)

cl_con = np.load(folder + 'Fig.5/' + 'clst_con307.npy')
cl_ch = np.load(folder + 'Fig.5/' + 'clst_ch397.npy')
hist5_1 = np.load(folder + 'Fig.5/' + 'L5hist_control.npy')
hist5_2 = np.load(folder + 'Fig.5/' + 'L5hist_chrimson.npy')
hist23_1 = np.load(folder + 'Fig.5/' + 'L23hist_control.npy')
hist23_2 = np.load(folder + 'Fig.5/' + 'L23hist_chrimson.npy')
```

```python
#Fig. 5f
titles = ['Control', 'Chrimson']
sns.set()
cm = 1/2.54
sns.set_style('ticks')
fsize = 6
spines = 0.5
fig = plt.figure(figsize=(9.5*cm, 4.5*cm))
color_con = ["black","black","black"]
color_chr = ["red","red","red"]
linestyle=["dashed","solid","dotted"]

gs =2
color_all = [color_con, color_chr]
frame = 100

p_val = []  
xlabels = np.arange(int(-frame*50/1000),int(frame*50/1000)*2+1,1)
ylabels =np.arange(0,4,1)
for s in range(2):
    ax = fig.add_subplot(1,2,s+1)
    fig.subplots_adjust(wspace=0.3)
    color = color_all[s]
    for i in range(3):
        y_F_dff=all_Fres[i+3*s]
        y_F_dff=np.array(y_F_dff)
        gauss = gaussian_filter1d(y_F_dff, sigma = gs)
        SEM = stats.sem(y_F_dff)
        time = np.arange(0,frame*3,1)
        y_mean = gauss.mean(axis = 0)
        ax.axvline(x=frame, ymin=0, ymax=300,color="black",lw = 0.5)
        ax.axvline(x=frame+20, ymin=0, ymax=300,color="black",lw = 0.5,linestyle ='--')
        ax.fill_between(time, y_mean-SEM,y_mean+SEM,color=color[i], alpha = 0.1)
        ax.plot(y_mean, color= color[i], linewidth=1,linestyle=linestyle[i])
        ax.set_xticks(range(0,220,20))
        ax.set_yticks(np.arange(0,0.035,0.01))
        plt.xlim(90,200)
        plt.ylim(-0.006,0.032)
        ax.set_xticklabels(xlabels,fontsize=fsize )
        ax.set_yticklabels(ylabels,fontsize=fsize )
        plt.title(titles[s], fontsize=fsize)
        ax.set_xlabel('Time (s)', size = fsize )
        ax.set_ylabel('dF/F0 (%)', size = fsize )
        ax.spines['right'].set_visible(False)
        ax.spines['top'].set_visible(False)
        ax.spines["bottom"].set_linewidth(spines)
        ax.spines["left"].set_linewidth(spines)
        ax.yaxis.set_ticks_position('left')
        ax.tick_params(direction = 'inout', length = 3,pad = 1,width = 0.5)
```
