```python
#import
import pickle
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
from scipy import stats
from scipy.ndimage import gaussian_filter1d
import seaborn as sns
```

```python
# Load data
folder = '/Volumes/Elements/data_for_uploaded/'

# Load pickle file
with open(folder + 'Fig.S12/' + 'dFovF.txt', "rb") as f:
    Fdd_all = pickle.load(f)

# Load CSV files
df_paw = pd.read_csv(folder + 'Fig.S12/' + 'FigS12c_withdraw.csv')
df = pd.read_csv(folder + 'Fig.S12/' + 'df_forFigS12.csv')
df2 = pd.read_csv(folder + 'Fig.S12/' + 'df2_forFigS12.csv')
```

```python
#FigS12c
paw1 =(df_paw[(df_paw['chrimson']==True)&(df_paw['OP']==1)])['Paw_withdraw']
paw2 =(df_paw[(df_paw['chrimson']==True)&(df_paw['OP']==2)])['Paw_withdraw']
paw3 =(df_paw[(df_paw['chrimson']==True)&(df_paw['OP']==3)])['Paw_withdraw']
paw1con =(df_paw[(df_paw['chrimson']==False)&(df_paw['OP']==1)])['Paw_withdraw']
paw2con=(df_paw[(df_paw['chrimson']==False)&(df_paw['OP']==2)])['Paw_withdraw']
paw3con =(df_paw[(df_paw['chrimson']==False)&(df_paw['OP']==3)])['Paw_withdraw']

sns.set()
cm = 1/2.54
sns.set_style('ticks')
fsize = 5
spines = 0.5

fig = plt.figure(figsize=(5*cm, 3*cm))
ax = fig.add_subplot(1, 2, 1)
plt.title('Control', fontsize=fsize)
sns.despine()
y_mean = np.array([paw1con,paw2con,paw3con]).mean(axis = 1)
SEM = stats.sem(np.array([paw1con,paw2con,paw3con]),axis = 1)

plt.errorbar(np.arange(0,3),y_mean,SEM, marker = 'o',color = 'black',markersize=2,lw = 1)
plt.ylim(0,35)
plt.xticks([-0.2, 0,1,2])
ax.set_xticklabels([None,1,2,3],fontsize=fsize)
ax.set_yticks([0,20,40])
ax.set_yticklabels([0,20,40],fontsize=fsize)
ax.set_xlabel('Session', size = fsize)
ax.set_ylabel('Paw withdrawal freq (%)', size = fsize)
ax.spines["bottom"].set_linewidth(spines)
ax.spines["left"].set_linewidth(spines)
ax.tick_params(direction = 'inout', length = 4,pad = 1,width = 0.5)

y_mean = np.array([paw1,paw2,paw3]).mean(axis = 1)
SEM = stats.sem(np.array([paw1,paw2,paw3]),axis = 1)

ax = fig.add_subplot(1, 2, 2)
plt.title('ChrimsonR', fontsize=fsize)
sns.despine()
plt.errorbar(np.arange(0,3),y_mean,SEM, marker = 'o',color = 'red',markersize=2,lw = 1)
plt.ylim(0,35)
plt.xticks([-0.2, 0,1,2])
ax.set_xticklabels([None,1,2,3],fontsize=fsize)
ax.set_yticks([0,20,40])
ax.set_yticklabels([0,20,40],fontsize=fsize)
ax.set_xlabel('Session', size = fsize)
ax.spines["bottom"].set_linewidth(spines)
ax.spines["left"].set_linewidth(spines)

ax.tick_params(direction = 'inout', length = 4,pad = 1,width = 0.5)
```

```python

gs=5
y=[np.array(Fdd_all[i]).mean(axis = 1) for i in range(4)]
g=[gaussian_filter1d(y[i], sigma = gs) for i in range(4)]

sns.set()
cm = 1/2.54
sns.set_style('ticks')
fsize = 7
spines = 0.5
time = np.arange(0,300,1)
fig = plt.figure(figsize=(9*cm, 4.5*cm))
fig.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=0.3, hspace=0.4)
color=['black','grey',"red","lightpink"]
color2=['black','lightgrey',"red","lightpink"]
gs =5
xtick = np.arange(0,320,20)
xlab = np.arange(-5,11,1)
ylim = [-0.02,0.042]
yltick= np.arange(-2,6,1)/100
ylab= np.arange(-2,6,1)
for s in range(2):
    ax = fig.add_subplot(1, 2, 1+s)
    for i in range(2):
        plt.plot(g[i+2*s][0],color=color[i+2*s], linewidth=1)
        SEM = stats.sem(np.array(Fdd_all[i+2*s])[0])
        plt.fill_between(time, g[i+2*s][0]-SEM,g[i+2*s][0]+SEM,color=color2[i+2*s], alpha = 0.1)
        ax.axes.xaxis.set_ticks(xtick )   
        ax.set_xticklabels(xlab,fontsize=fsize )
        ax.set(xlim=[90,235])
        ax.set_yticks(yltick)
        ax.set_yticklabels(ylab,fontsize=fsize )
        ax.set(ylim=ylim)
        ax.tick_params(direction = 'inout', length = 3,pad = 1,width = 0.5)
        ax.spines["bottom"].set_linewidth(spines)
        ax.spines["left"].set_linewidth(spines)
        sns.despine()
```

```python
#show AUC with vs without withdrawal
yticks = np.arange(-10,30,5)
cm = 1/2.54
fsize = 7
ylim1 = [-16,23]
spines = 0.5
fig = plt.figure(figsize=(8*cm, 3.5*cm))
labels = np.arange(-30,50,10)
color_ed=['darkgray','black']*3+['lightcoral','red']*3
palette=[{False:'grey',True:'black'},{False:'lightpink',True:'red'}]
boxprops = dict(linewidth=1)
for ch,h in zip(['control','Chrimson'],range(2)):
    ax = fig.add_subplot(1, 2, h+1)
    plt.title(ch,size=fsize)
    sns.despine()
    sns.boxplot(x="session", y='auc_0-5s', data=df[(df['chrimson']==ch)],hue = 'withdrawal',width=0.5,showfliers=False,ax=ax,medianprops=dict(color="white",linewidth=1),linewidth=0.5,boxprops = boxprops,palette=palette[h])
    ax.set_xlabel('Session', size = fsize)
    ax.set_ylabel('AUC', size = fsize)
    ax.set(ylim=ylim1)
    ax.set_xticklabels([1,2,3],fontsize=fsize)
    ax.set_yticks(yticks)
    ax.set_yticklabels(yticks,fontsize=fsize)
    ax.tick_params(direction = 'inout', length = 3,pad = 1,width = 0.5)
    ax.spines["bottom"].set_linewidth(spines)
    ax.spines["left"].set_linewidth(spines)
    ax.legend().remove()
    for b,box in enumerate(ax.artists):
        box.set_edgecolor(color_ed[b+6*h])
```

```python
#show AUC in the three sessions
sns.set()
cm = 1/2.54
sns.set_style('ticks')
fsize = 7
spines = 0.5
ylim=[[-1.2,1.5],[-1.8,2.3]]
ytick=[np.arange(-1,2,1),np.arange(-1,3,1)]
fig = plt.figure(figsize=(10*cm, 3.5*cm))
fig.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=0.6, hspace=0.4)
color=['grey',"lightpink",'black',"red"]
color2=['darkgray',"lightcoral",'black',"red"]
medianprops=dict(color="white",linewidth=1)
paw=['-no withdr','-withdr']
for wit,i in zip([False,True],range(2)):
    for ch,h in zip(['control','Chrimson'],range(2)):
        boxprops = dict(linewidth=1,edgecolor=color2[h+i*2])
        ax = fig.add_subplot(1, 4,h+i*2+1)
        plt.title(ch+paw[i],size=fsize)
        sns.boxplot(x="session", y='auc', data=df[(df['chrimson']==ch)&(df['withdrawal']==wit)],width=0.4,showfliers=False,ax=ax,color=color[h+i*2],medianprops=medianprops,linewidth=0.5
                ,boxprops=boxprops  )
        ax.set_xlabel('Session', size = fsize)
        ax.set(ylim=ylim[i])
        ax.set_xticklabels([1,2,3],fontsize=fsize)
        ax.set_yticks(ytick[i])
        ax.set_yticklabels(ytick[i],fontsize=fsize)
        ax.tick_params(direction = 'inout', length = 3,pad = 1,width = 0.5)
        ax.spines["bottom"].set_linewidth(spines)
        ax.spines["left"].set_linewidth(spines)
        ax.legend().remove()
        ax.set_ylabel(None)
        sns.despine()
```
